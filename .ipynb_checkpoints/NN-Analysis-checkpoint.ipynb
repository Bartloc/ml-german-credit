```python
X = data_num.iloc[:, :-5]
y=data_num.iloc[:,-1]
X.head(5)
```

```python
y.head(5)
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20) 
X_train , y_train
X_test.size
y_test.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler()  
scaler.fit(X_train)

X_train = scaler.transform(X_train)  
X_test = scaler.transform(X_test) 
```

```python
scaler = StandardScaler()  
scaler.fit(X)

X_scaled = scaler.transform(X)  
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
pca.fit(X_scaled)
x_pca=pca.transform(X_scaled)
x_pca.shape,X.shape
```

```python
plt.scatter(x_pca[:,0],x_pca[:,1],c=y)
```

```python
components=pd.DataFrame(pca.components_,columns=names[:-5])
components
sb.heatmap(components,cmap='plasma')
```

```python
from sklearn.neural_network import MLPClassifier 
mlp = MLPClassifier(hidden_layer_sizes=(100,40,100), max_iter=1000,verbose=True)  
mlp.fit(X_train, y_train.values.reshape(-1))  
```

```python
predictions = mlp.predict(X_test) 
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
#a,b,c,d,=confusion_matrix(y_test,predictions)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=300)
rfc.fit(X_train,y_train)
```

```python
predForest=rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test,predForest))
print(classification_report(y_test,predForest))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier()
dtree.fit(X_train,y_train)
```

```python
predictions=dtree.predict(X_test)
```

```python
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
```
